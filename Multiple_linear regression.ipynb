CREATING MY OWN MULTIPLE LR

```python
class MyLR:
  def __init__(self):
    self.coef_ = None
    self.intercept_ = None

  def fit(self, X_train, Y_train):

    if isinstance(X_train, pd.DataFrame):
        X_train_values = X_train.values
    else:
        X_train_values = X_train

    X_train_augmented = np.insert(X_train_values, 0, 1, axis=1)


    betas = np.linalg.inv(np.dot(X_train_augmented.T, X_train_augmented)).dot(X_train_augmented.T).dot(Y_train)

    self.intercept_ = betas[0]
    self.coef_ = betas[1:]

  def predict(self, X_test):

    if isinstance(X_test, pd.DataFrame):
        X_test_values = X_test.values
    else:
        X_test_values = X_test

    Y_pred = np.dot(X_test_values, self.coef_) + self.intercept_
    return Y_pred


model = MyLR()
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

print(Y_pred)
from sklearn.metrics import r2_score
print(r2_score(Y_test, Y_pred))

result = model.predict(np.array([[2, 2600, 7, 5]]))
print("Expected Body Result=", result[0])
```
