UNIVARIENT ANALYSIS

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("students_info.csv")
print(data.head())
```

```python
print("Dataset Info:")
print(data.info())
```

```python
print("Statistical Summary:")
print(data.describe())
```

```python
print("Midterm Analysis")
print("Mean:", data["Midterm"].mean())
print("Median:", data["Midterm"].median())
print("Max:", data["Midterm"].max())
print("Min:", data["Midterm"].min())
```

```python
print("Endterm Analysis")
print("Mean:", data["Endterm"].mean())
print("Median:", data["Endterm"].median())
print("Max:", data["Endterm"].max())
print("Min:", data["Endterm"].min())
```

Histogram for midterm

```python
data["Midterm"].hist()
plt.title("Midterm Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

Histogram for end term

```python
data["Endterm"].hist()
plt.title("Endterm Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```

Pie chart for courses

```python
import matplotlib.pyplot as plt

course_counts = data["Course"].value_counts()

plt.pie(course_counts, labels=course_counts.index, autopct="%1.1f%%")
plt.title("Course Distribution")
plt.show()
```

Pie chart for study level of students

```python
study_counts = data["Study_Level"].value_counts()

plt.pie(study_counts, labels=study_counts.index, autopct="%1.1f%%")
plt.title("Study Level Distribution")
plt.show()
```
