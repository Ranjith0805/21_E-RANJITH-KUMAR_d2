```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('student.csv')
```

```python

df.head()
```

```python
plt.scatter(df['12th Marks'],df['CGPA'])
plt.xlabel('12th marks')
plt.ylabel('CGPA')
plt.show()
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python

X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
X_train_processed = X_train.drop(columns=['Name'])
lr.fit(X_train_processed, y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[0].drop('Name').values.reshape(1,1))
```

```python
plt.scatter(df['12th Marks'],df['CGPA'])
plt.plot(X_train_processed['ID'],lr.predict(X_train_processed),color='red')
plt.xlabel('12th Marks')
plt.ylabel('CGPA')
```

```python
m = lr.coef_
m
```

```python
b = lr.intercept_
b
```
