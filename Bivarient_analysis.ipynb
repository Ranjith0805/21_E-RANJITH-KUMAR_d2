BI_VARIENT ANALYSIS

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("students_info.csv")
data.head()
```

Scatter plot btw midterm and endterm

```python
data.plot.scatter(x="Midterm", y="Endterm")
plt.title("Midterm vs Endterm Marks")
plt.show()
```

Co-Relation btw midterm and end term

ex: if >0.7--strong,0.3-0.7--medium,<0.3--weak

```python
corr = data["Midterm"].corr(data["Endterm"])
print("Correlation:", corr)
```

Course Vs Average marks

```python
course_avg = data.groupby("Course")[["Midterm","Endterm"]].mean()
print(course_avg)
```

```python
study_avg = data.groupby("Study_Level")[["Midterm","Endterm"]].mean()
print(study_avg)
```

Course vs Average marks (Bar Graph)

```python
course_avg = data.groupby("Course")[["Midterm","Endterm"]].mean()
course_avg.plot(kind="bar")

plt.title("Average Marks by Course")
plt.ylabel("Marks")
plt.show()
```

Study level vs Average marks (Bar Graph)

```python
study_avg = data.groupby("Study_Level")[["Midterm","Endterm"]].mean()
study_avg.plot(kind="bar")

plt.title("Average Marks by Study Level")
plt.ylabel("Marks")
plt.show()
```
